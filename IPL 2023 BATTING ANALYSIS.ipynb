## HIGHEST RUN SCORER

```python
runs_made = df.groupby(['PLAYER'],as_index=False)['RUNS'].sum().sort_values(by='RUNS', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(22,7)})
sns.barplot(x = 'PLAYER', y= 'RUNS',data= runs_made,palette='Blues_r')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 15,rotation = 0)
plt.yticks(fontsize = 30)
plt.xlabel("PLAYERS",fontsize = 35)
plt.ylabel("RUNS",fontsize = 35)
plt.title("HIGHEST RUN SCORER",fontsize = 36)
plt.show()
```

### FROM THE ABOVE GRAPH WE CAN SEE THAT SHUBMAN GILL IS THE HIGHEST RUN SCORER IN THE TATA IPL 2023 SEASON

#    

# HIGHEST STRIKE RATE

```python
strike_rate = df.groupby(['PLAYER'],as_index=False)['STRIKE RATE'].sum().sort_values(by='STRIKE RATE', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(22,7)})
sns.barplot(x = 'PLAYER', y= 'STRIKE RATE',data= strike_rate,palette = 'BrBG')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 15,rotation = 0)
plt.yticks(fontsize = 30)
plt.xlabel("PLAYERS",fontsize = 35)
plt.ylabel("STRIKE RATE",fontsize = 35)
plt.title("HIGHEST STRIKE-RATE",fontsize = 36)
plt.show()
```

### FROM THE ABOVE GRAPH WE CAN SEE THAT RASHID KHAN HAS THE HIGHEST STRIKE RATE

# MOST FOURS

```python
four = df.groupby(['PLAYER'],as_index=False)['4s'].sum().sort_values(by='4s', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(22,7)})
sns.barplot(x = 'PLAYER', y= '4s',data= four,palette='gist_heat')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 15,rotation = 0)
plt.yticks(fontsize = 30)
plt.xlabel("PLAYERS",fontsize = 35)
plt.ylabel("FOURS",fontsize = 35)
plt.title("MOST FOURS",fontsize = 36)
plt.show()
```

### FROM THE ABOVE GRAPH SHUBMAN GILL LEADS WITH THE MOST FOURS

# MOST SIXES

# TOTAL TEAM RUNS
