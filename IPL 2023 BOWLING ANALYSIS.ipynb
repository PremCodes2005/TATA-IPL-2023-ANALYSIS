## HIGHEST WICKET TAKER

```python
wickets_taken = df.groupby(['PLAYER'],as_index=False)['WICKETS'].sum().sort_values(by='WICKETS', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(22,7)})
sns.barplot(x = 'PLAYER', y= 'WICKETS',data= wickets_taken,palette='PuOr')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 15,rotation = 0)
plt.yticks(fontsize = 30)
plt.xlabel("PLAYERS",fontsize = 35)
plt.ylabel("WICKETS",fontsize = 35)
plt.title("HIGHEST WICKET TAKER",fontsize = 36)
plt.show()
```

### FROM THE ABOVE GRAPH WE CAN SEE THAT MOHAMMED SHAMI IS THE HIGHEST WICKET TAKER IN THE TATA IPL 2023 SEASON

#    

# BEST BOWLING AVERAGE

```python
b_avg= df.groupby(['PLAYER'],as_index=False)['AVERAGE'].sum().sort_values(by='AVERAGE', ascending=False).tail(10) 
sns.set(rc={'figure.figsize':(22,7)})
sns.barplot(x = 'PLAYER', y= 'AVERAGE',data= b_avg,palette = 'BrBG')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 15,rotation = 0)
plt.yticks(fontsize = 30)
plt.xlabel("PLAYERS",fontsize = 35)
plt.ylabel("BOWLING AVERAGE",fontsize = 35)
plt.title("BEST BOWLING AVERAGE",fontsize = 36)
plt.show()
```

### FROM THE ABOVE GRAPH WE CAN SEE THAT MARK WOOD HAS THE BEST BOWLING AVERAGE

#        



# LOWEST ECONOMY

```python
eco = df.groupby(['PLAYER'],as_index=False)['ECONOMY'].sum().sort_values(by='ECONOMY', ascending=False).tail(10) 
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x = 'PLAYER', y= 'ECONOMY',data= eco,palette='gist_heat')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 15,rotation = 0)
plt.yticks(fontsize = 30)
plt.xlabel("PLAYERS",fontsize = 35)
plt.ylabel("ECONOMY",fontsize = 35)
plt.title("BOWLING ECONOMY",fontsize = 36)
plt.show()
```

### FROM THE ABOVE GRAPH P . DUBEY HAS THE LOWEST ECONOMY

# MOST MAIDEN OVERS

```python
md = df.groupby(['PLAYER'],as_index=False)['MAIDENS'].sum().sort_values(by='MAIDENS', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(21,7)})
sns.barplot(x = 'PLAYER', y= 'MAIDENS',data=md,palette='afmhot')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 15,rotation = 0)
plt.yticks(fontsize = 30)
plt.xlabel("PLAYERS",fontsize = 35)
plt.ylabel("OVERS",fontsize = 35)
plt.title("MAIDEN OVERS",fontsize = 36)
plt.show()
```

### FROM THE ABOVE GRAPH WE CAN SEE THAT TRENT BOULT HAS THE MOST NUMBER OF MAIDEN OVERS
